```python
#!pip install opencv-python
import cv2
cv2.__version__
```

```python
img_path = 'images/image_1.png'
img_np = cv2.imread(img_path)
# when cv2 reads an image, it reads as a numpy ndimentional array
print(type(img_np))
print('images shape', img_np.shape)
```

```python
from matplotlib import pyplot as plt
#ax = plt.axes()
#ax.set_axis_off()
plt.imshow(img_np)
plt.show()
```

```python
cv2.imwrite('images/my_saved_img.png', img_np)
```

```python
# crop image using numpy slice
plt.imshow(img_np[30:200, 100:280])
```

```python
# crop image using numpy slice
plt.imshow(img_np[225:325, 50:350])
```

```python
# access BRG pixels\n
img_np[0][0]
#img_np[0][0][0]
#img_np[0][0][1]
#img_np[0][0][2]

#img_np[100][275]
```

```python
# access only blue color
blue = img_np[100,100,0]
print('blue', blue)
```

```python
# inverse image
plt.imshow(255 - img_np)
```

```python
negative_img = 255 - img_np
negative_img[0][0]
```

```python
cv2.imshow('image',negative_img)
k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('images/neg_out.jpg',negative_img)
    cv2.destroyAllWindows()
```
